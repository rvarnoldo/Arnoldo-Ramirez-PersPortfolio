```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
travel_1 = pd.read_csv('/travel_1.csv', index_col=0, encoding='latin-1')
travel_1
```

```python
travel_2 = pd.read_csv('/travel_2.csv')
travel_2
```

'Autobus' boxplot

```python
plt.style.use(['seaborn'])

plt.boxplot(x=travel_2['Autobus'])
plt.title('Entidad')
plt.xlabel('Autobus')
plt.show()
```

Creating a 3D Graph relating 'Automovil', 'Autobus', and 'Camioneta'

```python
plt.style.use('ggplot')
sns.set(style = 'darkgrid')

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = travel_2['Automovil']
y = travel_2['Autobus']
z = travel_2['Camioneta']

ax.set_xlabel('Automovil')
ax.set_ylabel('Autobus')
ax.set_zlabel('Camioneta')
plt.title('Gráfica tridimensional utilizando SEABORN')
ax.scatter(x, y, z, c='red')
ax.set_xticks(np.arange(0,x.max(), 200))
ax.set_yticks(np.arange(0,y.max(), 100))
ax.set_zticks(np.arange(0,z.max(), 50))
plt.show()
```

Building a Residual Graph from 3 Variables, on which 'Automobil' is the dependent variable. 'Autobus' and 'Camioneta' are independent variables.

```python
fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(121)
sns.residplot(data = travel_2, x=travel_2['Automovil'], y=travel_2['Autobus'], color='r', ax=ax1)
ax1.set_title('Modelo residual: Automovil vs Autobus')
ax1.set_xlabel('Automovil')
ax1.set_ylabel('Error en el ajuste')

ax2 = fig.add_subplot(122)
sns.residplot(data = travel_2, x=travel_2['Automovil'], y=travel_2['Camioneta'], color='r', ax=ax2)
ax2.set_title('Modelo residual: Automovil vs Camioneta')
ax2.set_xlabel('Automovil')
ax2.set_ylabel('Error en el ajuste')
```

Regression model with 'Automovil' and 'Autobus as variables

```python
sns.set_style('whitegrid')
sns.lmplot(data = travel_2, x='Automovil', y='Autobus')
plt.show()
```

kdeplot graph with relatoin between 'Automovil' and 'Autobuses' for the first 6 entities.

```python
travel_2_1to6 = travel_2[['Entidad','Automovil', 'Autobus']][:6]
travel_2_1to6
```

```python
sns.set(context='poster',
        style='white',
        palette='dark',
        font='sans-serif',
        font_scale=0.55,
        color_codes='r',
        rc=None)
sns.kdeplot(data = travel_2_1to6, x=travel_2_1to6['Automovil'], shade = True, color = 'Green')
plt.ylabel('Autobus')
plt.title('kdeplot')
```

Histogram that relates number of 'Automovil' and Entity

```python
sns.set(context="poster", 
        style="white", 
        palette="dark", 
        font="sans-serif", 
        font_scale=0.55, 
        color_codes='r', 
        rc=None)

plt.bar(data=travel_2_1to6, x=travel_2_1to6['Entidad'], height=travel_2_1to6['Automovil'])
plt.yticks(np.arange(0,travel_2_1to6['Automovil'].max()+10,50))
plt.xticks(rotation=30)
plt.xlabel('ENTIDAD', fontweight='bold')
plt.ylabel('AUTOMÓVILES', fontweight='bold')
plt.title('ENTIDAD VS AUTOMÓVILES', fontweight='bold')
```

Pie chart relating 'Automovil' and the first 6 entities in the table

```python
colores = ["#EE6052", "#60D396", "#AAF683", '#66b3ff','#99ff99','#ffcc99']
plt.pie(travel_2_1to6['Automovil'], labels=travel_2_1to6['Entidad'], autopct="%0.1f%%", colors=colores)
plt.axis('equal')
plt.show()
```
